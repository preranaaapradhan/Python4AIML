```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

pd.set_option('display.float_format', lambda x: '%3f' % x)

data = pd.read_csv('/Users/preranaapradhan/Documents/AIML/Python4AIML/Week3 - EDA/Melbourne_Housing.csv')
```

HISTOGRAM
- We use histplot() function of seaborn to create histograms.

```python
print(data.columns)
sns.histplot(data, x = 'Price')
```

```python
# Customization of a Histogram

plt.title('Histogram:Price')
plt.xlim(3000, 50000)
plt.ylim(0, 70)
plt.xlabel('Price of cars')
plt.ylabel('Frequency')
sns.histplot(data, x='Price', color='orange')
```

```python
sns.histplot(data, x='Price', kde='True')
```

BOXPLOTS

```python
g = sns.FacetGrid(data, col="body_style")
g.map(sns.histplot,"Price")
```

# BOXPLOT
It shows the distribution of numerical data and skewness through displaying the data quartiles.
Is also called five number summary plot where the five number summary is : minimum, first quartile, median, third  quartile and maximum.
boxplot() is the function used to create boxplot.
